```python
a = linspace(0, 2*pi, 21)
%precision 3
print(a)
print(a.reshape(3, 7))
```

```python
b = sin(a)
print(b)
print(b>0) # 会每个都会进行比较
```

```python
%matplotlib inline
plot(a,b) 
```

```python
mark = b > 0
print(mark)
print(a)
print(a[mark], b[mark])
# print(a & b> 0)
plot(a[mark], b[mark], 'bo')
```

```python
%matplotlib inline 
a = linspace(0, 2*pi, 50)
# linspace?
x = sin(a)
y = cos(a)
plot(x)
# hold()
plot(y)
# legend('sin(x) cos(x)')
# legend(['sin(x) cos(x)'])
legend(['sin', 'cos'])
title("sin&&cos")
```

```python
%matplotlib inline 
a = linspace(0,10,50)
y = a**2
plot(a,'r-^')
figure()
plot(a,y, 'b-o')
```
